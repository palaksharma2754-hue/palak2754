```python
# import seaborn as sns
import pandas as pd
import matplotlib .pyplot as plt
df=pd.read_csv("amazon.csv")
df=pd.read_csv("amazon.csv")
```

```python
# df["last category"]=df["category"].str.split("|").str[-1]
df["category"].value_counts().tail(10).plot(kind="barh")
plt.title("top 10 categories")
plt.xlabel("no. of products")
plt.ylabel("category")
# plt.xticks(rotation=45)
plt.show()
```

```python
plt.hist(df["rating"].dropna(), bins=10)

plt.title("Product Ratings")
plt.xlabel("Rating")
plt.ylabel("Count")
# plt.show()
plt.xticks(rotation=45)
plt.show()
# sns.histplot(df,"rating")
```

```python
plt.hist(df["actual_price"], bins=20)

plt.title("Actual Price Distribution")
plt.xlabel("Price")
plt.ylabel("Products")
plt.show()
```

```python
top = df.sort_values("rating", ascending=False).head(10)

plt.bar(top["product_name"], top["rating"])

plt.xticks(rotation=90)
plt.title("Top Rated Products")
plt.show()
```

```python
plt.hist(df["discount_percentage"], bins=10)

plt.title("Discount Percentage")
plt.xlabel("Discount %")
plt.ylabel("Products")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df["actual_price"], df["discount_price"], alpha=0.5)
plt.title("Actual Price vs Discount Price")
plt.xlabel("Actual Price")
plt.ylabel("Discount Price")
plt.show()
```

```python
import matplotlib.pyplot as plt

colors = ["red", "blue", "green", "orange", "purple",
          "pink", "cyan", "gold", "brown", "lime"]

df["category"].value_counts().head(10).plot(
    kind="bar",
    figsize=(10,6),
    color=colors,
    edgecolor="black"
)

plt.title("Top 10 Categories", fontsize=16)
plt.xlabel("Category", fontsize=12)
plt.ylabel("Number of Products", fontsize=12)
plt.xticks(rotation=45, ha="right")
plt.grid(axis="y", linestyle="--", alpha=0.5)

plt.show()
```
