```python
# import seaborn as sns
import pandas as pd
import matplotlib .pyplot as plt
df=pd.read_csv("amazon.csv")
df1=pd.read_csv("Amazon_c.csv")
```

```python
# import matplotlib.pyplot as plt

# Sirf last category name
df["category"] = df["category"].str.split("|").str[-1].str.strip()

# Top 10 Categories
top_categories = df["category"].value_counts().head(10)

# Different colors
colors = [
    "#FF6B6B", "#4ECDC4", "#45B7D1", "#FFA726", "#66BB6A",
    "#AB47BC", "#EC407A", "#29B6F6", "#FFCA28", "#8D6E63"
]

plt.figure(figsize=(12,6))

plt.bar(
    top_categories.index,
    top_categories.values,
    color=colors,
    edgecolor="black",
    linewidth=1.5
)

plt.title("Top 10 Product Categories", fontsize=18, fontweight="bold")
plt.xlabel("Category", fontsize=13)
plt.ylabel("Number of Products", fontsize=13)

plt.xticks(rotation=30, ha="right")
plt.grid(axis="y", linestyle="--", alpha=0.4)

# Values on bars
for i, value in enumerate(top_categories.values):
    plt.text(i, value + 1, str(value), ha="center", fontsize=10, fontweight="bold")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Top 10 highest rated products
top = df.sort_values("rating", ascending=False).head(10)

# Product names chhote karo
top["short_name"] = top["product_name"].str.slice(0, 30)

colors = [
    "#FF6B6B", "#4ECDC4", "#45B7D1", "#FFA726", "#66BB6A",
    "#AB47BC", "#EC407A", "#29B6F6", "#FFCA28", "#8D6E63"
]

plt.figure(figsize=(13,6))

bars = plt.bar(
    top["short_name"],
    top["rating"],
    color=colors,
    edgecolor="black"
)

plt.title("Top 10 Highest Rated Products", fontsize=18, fontweight="bold")
plt.xlabel("Product Name", fontsize=13)
plt.ylabel("Rating", fontsize=13)

plt.xticks(rotation=45, ha="right")
plt.grid(axis="y", linestyle="--", alpha=0.4)

# Rating values on bars
for bar in bars:
    plt.text(
        bar.get_x() + bar.get_width()/2,
        bar.get_height() + 0.02,
        round(bar.get_height(), 1),
        ha="center",
        fontsize=10,
        fontweight="bold"
    )

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Top 10 Highest Discount Products
top = df.sort_values("discount_percentage", ascending=False).head(10).copy()

# Product names short
top["product_name"] = top["product_name"].str[:25]

colors = ["#FF6B6B","#4ECDC4","#45B7D1","#FFA726","#66BB6A",
          "#AB47BC","#EC407A","#29B6F6","#FFCA28","#8D6E63"]

plt.figure(figsize=(13,6))

# Lollipop lines
plt.hlines(
    y=top["product_name"],
    xmin=0,
    xmax=top["discount_percentage"],
    color="gray",
    linewidth=2
)

# Colorful dots
plt.scatter(
    top["discount_percentage"],
    top["product_name"],
    s=220,
    color=colors,
    edgecolor="black"
)

plt.title("Top 10 Highest Discount Products", fontsize=18, fontweight="bold")
plt.xlabel("Discount Percentage (%)")
plt.ylabel("Product Name")
plt.grid(axis="x", linestyle="--", alpha=0.4)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Top 6 Improvement Areas
top6 = df1["Improvement_Areas"].value_counts().head(6)

labels = top6.index.tolist()
values = top6.values.tolist()

# Circle close karne ke liye
values += values[:1]

angles = np.linspace(0, 2*np.pi, len(labels), endpoint=False).tolist()
angles += angles[:1]

plt.figure(figsize=(8,8))
ax = plt.subplot(111, polar=True)

# Colorful filled area
ax.fill(angles, values, color="deepskyblue", alpha=0.35)

# Main line
ax.plot(angles, values, color="crimson", linewidth=3, marker="o", markersize=8)

# Colorful points
colors = ["red", "orange", "gold", "limegreen", "blue", "purple"]
for angle, value, color in zip(angles[:-1], values[:-1], colors):
    ax.scatter(angle, value, color=color, s=120)

ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels, fontsize=10)

plt.title("Top Improvement Areas", fontsize=16, fontweight="bold", color="darkblue", pad=20)

plt.show()
```

```python
import pandas as pd

# Remove ₹ and commas, then convert to float
df['actual_price'] = (
    df['actual_price']
    .astype(str)
    .str.replace('₹', '', regex=False)
    .str.replace(',', '', regex=False)
)
df['actual_price'] = pd.to_numeric(df['actual_price'], errors='coerce')

df['discounted_price'] = (
    df['discounted_price']
    .astype(str)
    .str.replace('₹', '', regex=False)
    .str.replace(',', '', regex=False)
)
df['discounted_price'] = pd.to_numeric(df['discounted_price'], errors='coerce')

# Remove missing values
hex_df = df[['actual_price', 'discounted_price']].dropna()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))

plt.hexbin(
    hex_df['actual_price'],
    hex_df['discounted_price'],
    gridsize=25,
    cmap='viridis'
)

plt.colorbar(label='Products')
plt.xlabel("Actual Price")
plt.ylabel("Discounted Price")
plt.title("Actual Price vs Discounted Price Density")
plt.show()
```

```python
import matplotlib.pyplot as plt

cols = ['actual_price', 'discounted_price', 'discount_percentage', 'rating', 'rating_count']

corr = df[cols].corr()

plt.figure(figsize=(7,6))
plt.imshow(corr, cmap='coolwarm')

plt.xticks(range(len(cols)), cols, rotation=45)
plt.yticks(range(len(cols)), cols)

for i in range(len(cols)):
    for j in range(len(cols)):
        plt.text(j, i, f"{corr.iloc[i,j]:.2f}",
                 ha='center', va='center', color='black')

plt.colorbar()
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Actual Price
df['actual_price'] = (
    df['actual_price']
    .astype(str)
    .str.replace('₹', '', regex=False)
    .str.replace(',', '', regex=False)
)
df['actual_price'] = pd.to_numeric(df['actual_price'], errors='coerce')

# Discounted Price
df['discounted_price'] = (
    df['discounted_price']
    .astype(str)
    .str.replace('₹', '', regex=False)
    .str.replace(',', '', regex=False)
)
df['discounted_price'] = pd.to_numeric(df['discounted_price'], errors='coerce')

# Discount Percentage
df['discount_percentage'] = (
    df['discount_percentage']
    .astype(str)
    .str.replace('%', '', regex=False)
)
df['discount_percentage'] = pd.to_numeric(df['discount_percentage'], errors='coerce')

# Rating
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

# Rating Count
df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce')
```
